This notebook plots two neural networks with approxmiating the funcitons 

\begin{align}
y &= \sin(x) + \epsilon \\
    &= x^2 + \epsilon 
\end{align}

with $\epsilon \sim N(0, 0.001) $. The neural network has one hidden layer and 3 neurons (and a bias). The neuronal activitaion funciton is  sigmoid. 

```python
# Generate Paramters 

link = ['/home/fabian/Documents/DeepLearningTutorials/data/QuadraticRegression.npy', 
        '/home/fabian/Documents/DeepLearningTutorials/data/SineRegression.npy']

learning_rate = 0.0015
n_epochs = 1000
batch_size = 25
n_hidden = 3
L2_reg = 0.0001
```

```python
%matplotlib inline 
import numpy as np 
import matplotlib.pyplot as plt 
import sys
sys.path.append("/home/fabian/Documents/DeepLearningTutorials/code")
from mlp_extended import MLP_OPTIMIZATION
from linearRegression_sgd import LinearRegression
```

```python
for i in range(len(link)):
    opt = MLP_OPTIMIZATION(learning_rate = learning_rate, n_epochs = n_epochs, 
                          batch_size = batch_size, link = link[i],
                          n_hidden = n_hidden, L2_reg = L2_reg, 
                          n_out = 1, output_layer = LinearRegression )
    opt.predict()
```
